# Lecture 38 outlier treatment

# we use capping and flooring to handle outliers in n_hot_rooms

```python
np.percentile(df['n_hot_rooms'],[99])  #returns array
```

```python
#upper limit
uv = np.percentile(df['n_hot_rooms'],[99])[0]
```

```python
#get outliers
df[(df.n_hot_rooms > uv)]
```

```python
#cap outliers : impute 3*99%le at values greater than uv
df.n_hot_rooms[(df.n_hot_rooms > 3*uv)] = 3 * uv
```

```python
df[(df.n_hot_rooms > uv)]
```

# same for rainfall


```python
lv= np.percentile(df.rainfall,[1])[0] #lower value
```

```python
df[(df.rainfall < lv)]
```

```python
df.rainfall[(df.rainfall < 0.3*lv)] = 0.3*lv
```

```python
df[(df.rainfall < lv)]
```

# Ridge regression

```python
# first standardize the data
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(x_train)
```

```python
#transform x_train into scaled x_train
x_train_s = scaler.transform(x_train)
```

```python
x_test_s = scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r = Ridge(alpha = 0.5)  # here alpha is value ou our lambda
```

```python
lm_r.fit(x_train_s, y_train)
```

```python
r2_score(y_test, lm_r.predict(x_test_s)) 
```

```python
# find value of alpha/ lambda such that value of r2_score is maximum
# we have built in function validations curve for this
```

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
# check 100 lambda values in range of 10^-2 to 10^8
param_range = np.logspace(-2,8,100)
```

```python
param_range
```

```python
train_scores, test_scores = validation_curve(Ridge(), x_train_s, y_train, param_name="alpha", param_range=param_range, scoring="r2")
```

```python
print(train_scores)
```

```python
print(test_scores)
```

```python
train_mean = np.mean(train_scores, axis=1) # take mean of five r2 values for one lambda
```

```python
test_mean = np.mean(test_scores, axis=1)
```

```python
train_mean
```

```python
test_mean
```

```python
# max r2 value for test
max(test_mean)
```

```python
sns.jointplot(x=np.log(param_range), y= test_mean)
```

```python
# find lamda for max r2
np.where(test_mean==max(test_mean))
```

```python
param_range[31]
```

```python
lm_r_best = Ridge(alpha = param_range[31])
```

```python
lm_r_best.fit(x_train_s, y_train)
```

```python
r2_score(y_test, lm_r_best.predict(x_test_s))
```

```python
r2_score(y_train, lm_r_best.predict(x_train_s))
```
