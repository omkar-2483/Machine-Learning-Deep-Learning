# SVM based Clasification model

```python
df = pd.read_csv("./Movie_classification.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

# missing values

```python
df.Time_taken.fillna(value= df.Time_taken.mean(), inplace= True)
```

```python
df.info()
```

# dummy variables 

```python
df = pd.get_dummies(df, columns=["3D_available", "Genre"], drop_first=True, dtype=int)
```

```python
df.head()
```

# X - y split

```python
X = df.loc[:, df.columns!="Start_Tech_Oscar"]
```

```python
X.head()
```

```python
y = df["Start_Tech_Oscar"]
```

```python
y
```

```python
type(X)
```

```python
type(y)
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Standardize data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
```

```python
X_test_std =  sc.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

```python
# X_test_std.mean()
```

```python
# X_test_std[0].mean()
```

# SVM based classifier

```python
from sklearn import svm
```

```python
svm_clf = svm.SVC(kernel="linear", C=0.01)
```

```python
svm_clf.fit(X_train_std, y_train)
```

# predict values

```python
y_train_pred = svm_clf.predict(X_train_std)
```

```python
y_test_pred = svm_clf.predict(X_test_std)
```

```python
y_test_pred
```

# Model performance

```python
 from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
 # Number of support vectors for each class.
svm_clf.n_support_
```

# Grid Search 

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C':(0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000)}
```

```python
svm_clf_1= svm.SVC(kernel = "linear")
```

```python
svm_grid_lin = GridSearchCV(svm_clf_1, params, n_jobs=-1, cv=10, verbose=1, scoring="accuracy")
```

```python
svm_grid_lin.fit(X_train_std, y_train)
```

```python
svm_grid_lin.best_params_
```

```python
linsvm_clf = svm_grid_lin.best_estimator_
```

```python
confusion_matrix(y_test, linsvm_clf.predict(X_test_std))
```

```python
accuracy_score(y_test, linsvm_clf.predict(X_test_std))
```

# polynomial svm classifier

```python
svm_clf_p = svm.SVC(kernel= "poly", degree=2, C=0.1)
```

```python
svm_clf_p.fit(X_train_std, y_train)
```

```python
y_train_pred = svm_clf_p.predict(X_train_std)
y_test_pred = svm_clf_p.predict(X_test_std)
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
svm_clf_p.n_support_
```

# SVM classifier with radial kernel

```python
svm_clf_r = svm.SVC(kernel="rbf", gamma=0.5, C=10)
```

```python
svm_clf_r.fit(X_train_std, y_train)
```

```python
y_test_pred = svm_clf_r.predict(X_test_std)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
svm_clf_r.n_support_
```

# Grid search for radial kernel

```python
params = {'C': (0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50),
           'gamma':(0.001, 0.01, 0.1, 0.5, 1)}
```

```python
svm_clf_r = svm.SVC(kernel="rbf")
```

```python
svm_grid_r= GridSearchCV(svm_clf_r, params, n_jobs=-1, cv=3, verbose=1, scoring="accuracy")
```

```python
svm_grid_r.fit(X_train_std, y_train)
```

```python
svm_grid_r.best_params_
```

```python
radsvm_clf =svm_grid_r.best_estimator_
```

```python
accuracy_score(y_test, radsvm_clf.predict(X_test_std))
```
