# import libarries and data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Movie_classification.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

# missing values imputation

```python
# impute mean value of time_taken in all null
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value= df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

# Dummy Variables Creation
for categorical data

```python
df.head()
```

```python
df = pd.get_dummies(df, columns =[ "3D_available","Genre"], drop_first=True)
```

```python
df.head()
```

```python
df['3D_available_YES'] = df['3D_available_YES'].astype(int)
```

```python
df['Genre_Comedy'] = df['Genre_Comedy'].astype(int)
df['Genre_Drama'] = df['Genre_Drama'].astype(int)
df['Genre_Thriller'] = df['Genre_Thriller'].astype(int)
```

```python
df.head()
```

# X-Y split

```python
X = df.loc[: , df.columns != "Start_Tech_Oscar"]  #iloc takes two parameters rows and columns ":" for all rows
```

```python
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
Y = df['Start_Tech_Oscar']
```

```python
type(Y)
```

```python
Y.shape
```

```python
Y.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
X_train.head()
```

# Classification Tree
To predict if movie can win Oscar

Training
-

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth=3)  #shift + tab for helper
```

```python
clftree.fit(X_train, Y_train)
```

Prediction 
-

```python
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
```

```python
y_train_pred
```

Model Performance
-

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
confusion_matrix(Y_train, y_train_pred)
```

```python
confusion_matrix(Y_test, y_test_pred)
```

```python
accuracy_score(Y_test, y_test_pred)
```

Plotting decision tree
-

```python
dot_data = tree.export_graphviz(clftree, out_file=None, 
                           filled=True,feature_names=X_train.columns)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
Image(graph.create_png())
```

# Controlling Tree Growth

Minimum observations at leaf node
-

```python
clftree1 = tree.DecisionTreeClassifier(min_samples_leaf = 20)
clftree1.fit(X_train, Y_train)
dot_data = tree.export_graphviz(clftree1, out_file = None, feature_names=X_train.columns, filled=True)
graph1=pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

Multiple conditions
-

```python
clftree2 = tree.DecisionTreeClassifier(min_samples_leaf = 20, max_depth=4)
clftree2.fit(X_train, Y_train)
dot_data = tree.export_graphviz(clftree2, out_file = None, feature_names=X_train.columns, filled=True)
graph2=pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

# Bagging
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html

class sklearn.ensemble.BaggingClassifier(estimator=None, n_estimators=10, *, max_samples=1.0, max_features=1.0, bootstrap=True, bootstrap_features=False, oob_score=False, warm_start=False, n_jobs=None, random_state=None, verbose=0)

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(estimator = clftree, n_estimators=1000, bootstrap=True, n_jobs=-1, random_state=42) 
#n_estimators=we want 1000 trees
#bootstrap = take subset for each of 1000 trees with replacement
```

```python
bag_clf.fit(X_train, Y_train)
```

```python
confusion_matrix(Y_test, bag_clf.predict(X_test))
```

```python
accuracy_score(Y_test, bag_clf.predict(X_test))
```

# Random Forest
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, random_state=42)
```

```python
rf_clf.fit(X_train, Y_train)
```

```python
confusion_matrix(Y_test, rf_clf.predict(X_test))
```

```python
accuracy_score(Y_test, rf_clf.predict(X_test))
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf= RandomForestClassifier(n_estimators=250 , random_state=42)
```

```python
params_grid = {"max_features": [4,5,6,7,8,9,10], 
               "min_samples_split": [2,3,10]}
```

```python
grid_search = GridSearchCV(rf_clf, params_grid, n_jobs=-1, cv=5, scoring="accuracy")
```

```python
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_params_
```

```python
#best model
cvrf_clf = grid_search.best_estimator_
```

```python
confusion_matrix(Y_test, cvrf_clf.predict(X_test))
```

```python
accuracy_score(Y_test, cvrf_clf.predict(X_test))
```
