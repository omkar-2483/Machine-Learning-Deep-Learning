# import libarries and data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Movie_regression.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

# missing values imputation

```python
# impute mean value of time_taken in all null
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value= df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

# Dummy Variables Creation
for categorical data

```python
df.head()
```

```python
df = pd.get_dummies(df, columns =[ "3D_available","Genre"], drop_first=True)
```

```python
df.head()
```

```python
df['3D_available_YES'] = df['3D_available_YES'].astype(int)
```

```python
df['Genre_Comedy'] = df['Genre_Comedy'].astype(int)
df['Genre_Drama'] = df['Genre_Drama'].astype(int)
df['Genre_Thriller'] = df['Genre_Thriller'].astype(int)
```

```python
df.head()
```

# X-Y split

```python
X = df.loc[: , df.columns != "Collection"]
```

```python
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
Y = df['Collection']
```

```python
type(Y)
```

```python
Y.shape
```

```python
Y.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
X_train.head()
```

# Regression Tree

Training
-

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
regtree.fit(X_train, Y_train)
```

Prediction 
-

```python
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

Model Performance
-

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(Y_test, y_test_pred)
```

```python
r2_score(Y_train, y_train_pred)
```

```python
r2_score(Y_test, y_test_pred)
```

Plotting decision tree
-

```python
dot_data = tree.export_graphviz(regtree, out_file=None, 
                           filled=True, rounded=True, 
                           special_characters=True)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
Image(graph.create_png())
```

# Controlling Tree Growth

Maximun number of level in tree
-

```python
regtree1 = tree.DecisionTreeRegressor( max_depth = 3)  #shift + tab for help window
regtree1.fit(X_train, Y_train)
dot_data = tree.export_graphviz(regtree1, out_file = None, feature_names=X_train.columns, filled=True)
graph1=pydotplus.graph_from_dot_data(dot_data)
Image(graph1.create_png())
```

Minimum obserations in internal node
-

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split = 40)  # for further split min sample size should be 40
regtree2.fit(X_train, Y_train)
dot_data = tree.export_graphviz(regtree2, out_file = None, feature_names=X_train.columns, filled=True)
graph2=pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

Minimum observations at leaf node
-

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf = 25)  # for further split min sample size of leaf should be 25
regtree3.fit(X_train, Y_train)
dot_data = tree.export_graphviz(regtree3, out_file = None, feature_names=X_train.columns, filled=True)
graph3=pydotplus.graph_from_dot_data(dot_data)
Image(graph3.create_png())
```

Multiple conditions
-

```python
regtree4 = tree.DecisionTreeRegressor(min_samples_leaf = 25, max_depth=4)
regtree4.fit(X_train, Y_train)
dot_data = tree.export_graphviz(regtree4, out_file = None, feature_names=X_train.columns, filled=True)
graph4=pydotplus.graph_from_dot_data(dot_data)
Image(graph4.create_png())
```
